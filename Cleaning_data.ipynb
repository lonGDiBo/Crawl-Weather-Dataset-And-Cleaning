- Data is clean

```python
columns_to_keep = ['Date Time', 'p (mbar)', 'T (degC)', 'VPmax (mbar)', 'VPdef (mbar)', 'sh (g/kg)', 'rho (g/m**3)', 'wv (m/s)']
df = df.filter(columns_to_keep)
```

```python
df.head(5)
```

- Rename Columns

```python
titles = [
    "Pressure",
    "Temperature",
    "Saturation_vapor_pressure",
    "Vapor_pressure_deficit",
    "Specific_humidity",
    "Airtight",
    "Wind_speed",
]

feature_keys = [
    "p (mbar)",
    "T (degC)",
    "rh (%)",
    "VPmax (mbar)",
    "VPdef (mbar)",
    "sh (g/kg)",
    "rho (g/m**3)",
    "wv (m/s)",
]

rename_dict = dict(zip(feature_keys, titles))
df = df.rename(columns=rename_dict)
```

```python
df.columns
```

```python
df.head()
```

```python
# Convert to datetime
df['Date Time'] = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')  
```

```python
df.head(5)
```

- Calculate the mean value for each day

```python
df = df.set_index('Date Time')
df_resampled = df.resample('D').mean().round(2) 
```

```python
df_resampled.head(5)
```

- Export file csv

```python
df_resampled.to_csv('Weather_dataset_Cleaning.csv', index=True)
```
